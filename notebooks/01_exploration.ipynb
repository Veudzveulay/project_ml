```python
# Cellule 1 : Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Configuration pour voir toutes les colonnes
pd.set_option('display.max_columns', None)

# Cellule 2 : Chargement
# Le chemin dpend d'o tu lances le notebook, mais en relatif :
TRAIN_PATH = '../data/train.csv'

df = pd.read_csv(TRAIN_PATH)
print(f"Dimensions du dataset : {df.shape}")
df.head()
```

```python
# Cellule 3 : Distribution de la target
plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=df)
plt.title('Distribution de la Target (0 vs 1)')
plt.show()

# Calcul du pourcentage exact
ratio = df['target'].value_counts(normalize=True) * 100
print(f"Non-Transaction (0) : {ratio[0]:.2f}%")
print(f"Transaction (1) : {ratio[1]:.2f}%")
```

```python
# Cellule 4 : Vrification des valeurs manquantes et types
missing_values = df.isnull().sum().sum()
print(f"Nombre total de valeurs manquantes : {missing_values}")

# Vrification des types de donnes
print("\nTypes de donnes :")
print(df.dtypes.value_counts())

# Aperu des statistiques descriptives
df.describe()
```

```python
# Cellule 5 : Corrlation avec la Target
# numeric_only=True force pandas  ignorer la colonne "ID_code"
correlations = df.corr(numeric_only=True)['target'].abs().sort_values(ascending=False)

# On retire la target elle-mme de la liste
correlations = correlations.drop('target')

print("Top 10 des features les plus corrles avec la target :")
print(correlations.head(10))

# Visualisation des 30 plus fortes corrlations
plt.figure(figsize=(10, 6))
correlations.head(30).plot(kind='bar')
plt.title('Top 30 des Features les plus corrles (Valeur absolue)')
plt.ylabel('Coefficient de corrlation absolu')
plt.show()
```

```python
# Cellule 6 : Distribution des features importantes
# On prend les 4 variables les plus corrles trouves juste avant
top_features = correlations.head(4).index.tolist()

plt.figure(figsize=(16, 10))

for i, feature in enumerate(top_features):
    plt.subplot(2, 2, i + 1)
    # Distribution pour Target = 0 (Pas de transaction)
    sns.kdeplot(df.loc[df['target'] == 0, feature], label='Target = 0', fill=True, alpha=0.3)
    # Distribution pour Target = 1 (Transaction)
    sns.kdeplot(df.loc[df['target'] == 1, feature], label='Target = 1', fill=True, alpha=0.3)
    
    plt.title(f'Distribution de {feature} selon la Target')
    plt.legend()

plt.tight_layout()
plt.show()
```
