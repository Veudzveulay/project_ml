```python
# Cellule 1 : imports
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import roc_auc_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import json

pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")
```

```python
# Cellule 2 : Chargement + preprocessing (idem notebook 2)
# Chargement du dataset
df = pd.read_csv("../data/train.csv")

# Séparation X / y
X = df.drop(columns=["target", "ID_code"])
y = df["target"]

# Train/Validation split
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Standardisation
scaler = joblib.load("../models/scaler.pkl")
X_train_scaled = scaler.transform(X_train)
X_val_scaled = scaler.transform(X_val)

# Conversion en DataFrame pour garder les noms de colonnes
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X.columns)
X_val_scaled = pd.DataFrame(X_val_scaled, columns=X.columns)

# SMOTE
#sm = SMOTE(random_state=42)
#X_train_bal, y_train_bal = sm.fit_resample(X_train_scaled, y_train)
#X_train_bal = pd.DataFrame(X_train_bal, columns=X.columns)
X_train_bal = X_train_scaled
y_train_bal = y_train

print("Préprocessing terminé.")
print("Taille train :", X_train_bal.shape)
print("Taille validation :", X_val_scaled.shape)
```

```python
# Cellule 3 : Définition des modèles 
models = {
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=8, min_samples_split=10, random_state=42, n_jobs=-1),
    "XGBoost": XGBClassifier(eval_metric="logloss", n_estimators=200, max_depth=4, learning_rate=0.1, n_jobs=-1, random_state=42),
    "LightGBM": LGBMClassifier(random_state=42)
}

models
```

```python
# Cellule 4 : Entrainement + comparaison des modèles
results = {}

for name, model in models.items():
    print(f"Entraînement de : {name}")
    model.fit(X_train_bal, y_train_bal)
    preds = model.predict_proba(X_val_scaled)[:, 1]
    auc = roc_auc_score(y_val, preds)
    results[name] = auc

results
```

```python
# Cellule 5 : GridSearch LightGBM
param_grid = {
    "num_leaves": [31],
    "learning_rate": [0.03],
    "n_estimators": [300]
}

lgbm = LGBMClassifier(
    random_state=42,
    n_jobs=1,
    force_col_wise=True
)

grid = GridSearchCV(
    lgbm,
    param_grid,
    scoring="roc_auc",
    cv=3,
    n_jobs=1
)

grid.fit(X_train_bal, y_train_bal)

print("Meilleurs paramètres :", grid.best_params_)
print("Score ROC-AUC CV :", grid.best_score_)
```

```python
# Cellule 6 : Entrainement final du meilleur modèle
best_model = grid.best_estimator_
best_model.fit(X_train_bal, y_train_bal)

print("Modèle final entraîné.")
```

```python
# Celulle 7 : Evaluation finale
preds = best_model.predict_proba(X_val_scaled)[:, 1]
auc = roc_auc_score(y_val, preds)

print(f"ROC-AUC final : {auc:.4f}")

# Confusion matrix
y_pred_class = (preds > 0.5).astype(int)
cm = confusion_matrix(y_val, y_pred_class)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Prédiction")
plt.ylabel("Réel")
plt.show()

# Classification report
print("\nClassification Report :")
print(classification_report(y_val, y_pred_class))
```

```python
# Cellule 8 : Explicabilité SHAP du modèle LightGBM
import shap

# 1. Initialisation de l'explainer SHAP
explainer = shap.TreeExplainer(best_model)

# 2. Calcul des valeurs SHAP sur un échantillon (pour éviter un calcul trop lourd)
X_sample = X_val_scaled.sample(500, random_state=42)  # 500 lignes suffisent
shap_values = explainer.shap_values(X_sample)

# 3. Summary plot (graphique principal)
shap.summary_plot(shap_values, X_sample, plot_type="dot")

# 4. Bar plot des features les plus importantes
shap.summary_plot(shap_values, X_sample, plot_type="bar")
```

```python
# Cellule 9 : Sauvegarde du modèle
joblib.dump(best_model, "../models/best_model.pkl")
print("Modèle sauvegardé dans ../models/best_model.pkl")
```

```python
# Cellule 10 : Importance des features
importances = best_model.feature_importances_
indices = np.argsort(importances)[-20:]

plt.figure(figsize=(10, 6))
plt.barh(range(len(indices)), importances[indices])
plt.yticks(range(len(indices)), X.columns[indices])
plt.title("Top 20 Feature Importances (LightGBM)")
plt.show()
```

```python
# Cellule 11 : Préparation de l'interface et export du modèle
# Fonction d'inférence (pour l'API)
def predict_api(input_dict):
    # Charger les objets
    model = joblib.load("../models/best_model.pkl")
    scaler = joblib.load("../models/scaler.pkl")
    with open("features.json", "r") as f:
        features = json.load(f)

    # Convertir en DataFrame
    df = pd.DataFrame([input_dict], columns=features)

    # Standardisation
    df_scaled = scaler.transform(df)

    # Prédiction
    proba = model.predict_proba(df_scaled)[0][1]

    return float(proba)

print("Fonction d'inférence prête pour l'API.")
```
